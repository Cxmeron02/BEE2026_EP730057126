## <div style = 'background-color:beige'> <u> **Data and Methodology** </u> </font>

• Sources of data:

i) Yahoo Finance ETF sector data
ii) Wikipedia S&P 500 companies data
iii) Slickchart S&P500 company and sector weights data

• Why ETFs are useful sector proxies

— Exchange-traded funds are investment funds comprised of various assets
— Hence, sector-specific ETFs are practical proxies for the performance of the respective broader industry since they hold diversified baskets of stocks within each sector.
— SPDR S&P 500 ETF Trust (SPY) tracks the S&P 500 index, hence acting as a useful proxy for the entire U.S. equity market. 
— SPY provides a consistent and practical benchmark for comparing individual sector ETF returns and trading volumes, to effectively evaluate sector performance and ETF activity relative to the broader market.

XLK = Technology, XLF = Financials, XLE = Energy, XLV = Healthcare SPY = Market

• Choice of date range for data collection (2010-2025)
• Annualisation: 252 trading days per annum (63 days per quarter) is the standard figure used in financial modelling; different raw data sets had different numbers of trading days.

## <div class="alert alert-danger"> <center> **Comparing risk and return between different sectors** </div>

Higher returns does not necessarily equate to a more attractive investment opportunity if accompanied by a disproportionately high level of risk and volatility. Long-standing models such as CAPM and theories such as Efficient Market Hypthesis, argue that investors are rational agents that design thair portfolio according to their own preferenced regarding risk and returns in order to maximise their own utility. By comparing annualised data for ETF sector returns and volatility, I evaluate whether riskier sectors indeed reward investors with higher returns. In doing so, we can critique the usefulness of traditonal financial models used to price risk such as CAPM, which espouses an equal risk adjusted rate of return for all assets in  equilibrium, in the context of modern financial markets

### <div style = 'background-color:ivory'> <u> *Limitations* </u> </div>

Although I have made some key findings with clear and decisive implications for investors, the scale of my research means there are also key limitations regarding my analysis to consider:


A) Limitations of ETFs as proxies for analysis: 

– Cannot make causal inferences regarding any market trends identified in data analysis.
– Exclude smaller firms not listed on the S&P500
– ETF trading volume should not be interpreted as a pure measure of passive investing, as it may reflect short-term speculation or institutional investment strategies.
– Even though the Technology ETF outperforms others and ETF trading activity rises over time, we cannot make causal inferences that increased passive investment caused the increased market concentration, as we have not incorporated other factors.

 B) Limitations regarding available data

- No inclusion of direct passive investment data, i.e. ETF assets under management, net fund flows and mutual fund ownership data. 
- No publicly available data from high-frequency trading firms, which are likely to be increasingly perpetuating greater market activity/trading volume. 
- Valuation measures, such as price-to-earnings ratios, to assess whether outperformance was driven by fundamentals or by investor speculation/trading volume
